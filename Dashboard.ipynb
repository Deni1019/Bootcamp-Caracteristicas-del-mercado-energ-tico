```python
import streamlit as st
import plotly.express as px
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime, timedelta
```

```python
# Crear datos sinteticos
np.random.seed(42)
fechas = pd.date_range(start='2023-01-01', end='2024-12-31', freq='D')
n_productos = ['laptop','mouse','teclado','monitor','auriculares']
regiones = ['Norte', 'Sur', 'Este', 'Oeste','Centro']
```

```python
data=[]
for fecha in fechas:
    for region in regiones:
        for _ in range(np.random.poisson(10)):
            data.append({
                'fecha': fecha,
                'producto': np.random.choice(n_productos),
                'region'  : np.random.choice(regiones),
                'cantidad': np.random.randint(1, 6),
                'precio-unitario': np.random.uniform(50, 1500),
                'vendedor': f'vendedor_{np.random.randint(1, 21)}'
                 })
```

```python
df =pd.DataFrame(data)
print(df)
df['venta_total'] = df['cantidad'] * df['precio-unitario']
print(df)
```

```python
print(df.describe())
```

```python
df_montly=df.groupby(df['fecha'].dt.to_period('M'))['venta_total'].sum().reset_index()
df_montly['fecha']= df_montly['fecha'].astype(str)
print(df_montly)
```

```python
fig_monthly=px.line(df_montly,x='fecha',y='venta_total',title='Ventas Mensuales')
fig_monthly.update_layout(xaxis_title='Fecha',yaxis_title='Ventas Totales')
fig_monthly.update_traces(line=dict(width=3))
fig_monthly.show()
```

```python
df_productos = df.groupby('producto')['venta_total'].sum().sort_values(ascending=False)
fig_productos = px.bar(df_productos, x=df_productos.index, y=df_productos.values)
fig_productos.update_layout(xaxis_title='Producto', yaxis_title='Ventas Totales')
fig_productos.show()
```

```python
df_regiones = df.groupby('region')['venta_total'].sum().reset_index()
fig_regiones = px.pie(df_regiones, values='venta_total', names='region', title='Distribución de Ventas por Región')
fig_regiones.show()
```

```python
df_corr = df[['cantidad', 'precio-unitario', 'venta_total']].corr()
fig_heatmap = px.imshow(df_corr, color_continuous_scale='RdBu', title='Matriz de Correlación', text_auto=True)
fig_heatmap.show()
```

```python
fig_dist = px.histogram(df, x='venta_total', nbins=100, title='Distribución de Ventas Totales')
fig_dist.update_layout(xaxis_title='Ventas Totales', yaxis_title='Frecuencia')
fig_dist.show()
```
